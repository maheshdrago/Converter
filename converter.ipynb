```python
class Lengths():
    def meters_to_feet(self,meter):
        mf=meter*3.2808
        return mf
    def feet_to_meters(self,feet):
        fm=feet/3.2808
        return fm
    def inches_to_cm(self,inch):
        ic=inch/0.39370
        return ic
    def cm_to_inches(self,cm):
        ci=cm*0.39370
        return ci
    def mm_to_inches(self,mm):
        mi=mm*0.039370
        return mi
    def inches_to_mm(self,inch):
        im=inch/0.039370
        return im
    def inches_to_feet(self,inch):
        itf=inch*0.083333
        return itf
    def feet_to_inches(self,feet):
        fi=feet*12.000
        return fi
```

```python
class Weights():
    def kg_to_pound(self,kg):
        kp=kg*2.2046
        return kp
    def pound_to_kg(self,pound):
        pk=pound/2.2046
        return pk
    def stones_to_kg(self,stone):
        sk=stone/0.15747
        return sk
    def ounces_to_grams(self,ounce):
        og=ounce/35.274
        return og
    def grams_to_ounces(self,gram):
        go=gram*0.035274
        return go
    def ounces_to_pounds(self,ounce):
        op=ounce* 0.062500
        return op
    def pound_to_ounces(self,pound):
        po=pound*16.000
        return po
    def kg_to_stone(self,kg):
        ks=kg*0.157
        return ks
    
```

```python
class Temperature_conversions():
    def celsius_to_fahrenheit(self,celsius):
        celfa=((celsius*9)/5)+32
        return str(celfa)+"℉"
    def fahrenheit_to_celsius(self,fahrenheit):
        face=(fahrenheit-32)/1.8000
        return str(face)+"℃"
    def celsius_to_kelvin(self,celsius):
        ceke=celsius+ 273.15
        return str(ceke)+"K"
    def fahrenheit_to_kelvin(self,fahrenheit):
        fake=(fahrenheit - 32) * 5/9 + 273.15
        return str(fake)+"K"
    def kelvin_to_fahrenheit(self,kelvin):
        kefa=(kelvin - 273.15) * 9/5 + 32
        return str(kefa)+"℉"
    def kelvin_to_celsius(self,kelvin):
        kece=kelvin- 273.15
        return str(kece)+"℃"
    def kelvin_to_kelvin(self,kelvin):
        kk=kelvin
        return str(kk)+"K"
    def celsius_to_celsius(self,celsius):
        cc=celsius
        return str(cc)+"℃"
    def fahrenheit_to_fahrenheit(self,fahrenheit):
        ff=fahrenheit
        return str(ff)+"℉"
       
```
